# Linear Discriminant Analysis (LDA)

1) Objective: LDA is a supervised dimensionality reduction technique. It seeks to find the linear combinations of features that best separate different classes in the dataset.
2) Difference from PCA: LDA considers class information when projecting the data onto a lower-dimensional space. It maximizes the distance between class means while minimizing the spread within each class.
3) Visualization: LDA is particularly useful for visualizing class separation in the reduced space.

# Principal Component Analysis (PCA)

1) Objective: PCA is an unsupervised dimensionality reduction technique. It focuses on finding the directions (principal components) along which the data varies the most.
2) Difference from LDA: PCA does not consider class labels during dimensionality reduction. It aims to capture the overall variance in the dataset.
3) Visualization: PCA is often used for general feature reduction and visualization of data variability.

```r
# Set seed for reproducibility
set.seed(123)

# Generate random indices for train and test split
indices <- sample(1:nrow(nhanes_data), 0.8 * nrow(nhanes_data))

# Create training and testing sets
train_data <- nhanes_data[indices, ]
test_data <- nhanes_data[-indices, ]
```

```r
# Fit LDA model
lda_model <- lda(age_group ~ ., data = train_data)

# Make predictions on the test set
lda_predictions <- predict(lda_model, newdata = test_data)

# Create confusion matrix manually
conf_matrix_lda <- table(lda_predictions$class, test_data$age_group)
print(conf_matrix_lda)

# Calculate accuracy
accuracy_lda <- sum(diag(conf_matrix_lda)) / sum(conf_matrix_lda)
cat("Accuracy:", accuracy_lda, "\n")
```

The confusion matrix you provided for the LDA model results in a classification accuracy of approximately 95.83%, which indicates that the model performed well in distinguishing between the "Adult" and "Senior" age groups. In the confusion matrix:

- True Positives (Adult): 364
- True Negatives (Senior): 73
- False Positives: 8
- False Negatives: 11

The high accuracy suggests that the LDA model effectively captured the underlying patterns in the dataset related to age groups. The small number of misclassifications (8 cases of Adult misclassified as Senior and 11 cases of Senior misclassified as Adult) indicates a strong overall performance. However, it's essential to consider other metrics such as precision, recall, and F1 score, especially when dealing with imbalanced datasets. Further analysis and potentially fine-tuning the model could enhance its performance, but the initial results are promising for distinguishing between adult and senior age groups based on the given features.

```r

nhanes_data$age_group <- as.factor(nhanes_data$age_group)

# Extract features for PCA
pca_data <- nhanes_data[, -which(names(nhanes_data) == "age_group")]

# Standardize the data
pca_data_std <- scale(pca_data)

# Perform PCA
pca_model <- prcomp(pca_data_std, center = TRUE, scale. = TRUE)

# Extract the results
pca_results <- summary(pca_model)

# Plot scree plot
plot(1:length(pca_results$sdev), pca_results$sdev^2, type = "b", 
     xlab = "Principal Component", ylab = "Variance Explained", 
     main = "Scree Plot")

# Visualize the first two principal components
biplot(pca_model, scale = 0)
```

# Machine learning

```r
# Load necessary libraries
library(e1071)
library(pROC)

nhanes_data$age_group <- as.factor(nhanes_data$age_group)

nhanes_data_selected <- nhanes_data

# Split the data into training and testing sets
set.seed(123)
indices <- sample(1:nrow(nhanes_data_selected), 0.8 * nrow(nhanes_data_selected))
train_data <- nhanes_data_selected[indices, ]
test_data <- nhanes_data_selected[-indices, ]

# Train SVM model
svm_model <- svm(age_group ~ ., data = train_data)

# Make predictions on the test set
svm_predictions <- predict(svm_model, newdata = test_data)

# Evaluate performance
conf_matrix_svm <- table(svm_predictions, test_data$age_group)
accuracy_svm <- sum(diag(conf_matrix_svm)) / sum(conf_matrix_svm)

# Print confusion matrix and accuracy
print(conf_matrix_svm)
cat("Accuracy:", accuracy_svm, "\n")

# ROC curve
svm_roc <- roc(test_data$age_group, as.numeric(svm_predictions))
plot(svm_roc, main = "ROC Curve for SVM", col = "blue", lwd = 2)

# K-fold Cross-Validation
set.seed(123)
folds <- cut(seq(1, nrow(nhanes_data_selected)), breaks = 5, labels = FALSE)
cv_svm <- lapply(1:5, function(i) {
  test_indices <- which(folds == i, arr.ind = TRUE)
  test_data <- nhanes_data_selected[test_indices, ]
  train_data <- nhanes_data_selected[-test_indices, ]
  
  svm_model <- svm(age_group ~ ., data = train_data)
  svm_predictions <- predict(svm_model, newdata = test_data)
  
  conf_matrix <- table(svm_predictions, test_data$age_group)
  accuracy <- sum(diag(conf_matrix)) / sum(conf_matrix)
  
  return(accuracy)
})

# Print cross-validated results
cat("Cross-validated Accuracy:", mean(unlist(cv_svm)), "\n")
```

The confusion matrix for the Support Vector Machine (SVM) model shows that it performed exceptionally well on the provided dataset. With an accuracy of approximately 97.37%, the model correctly classified 374 instances of the "Adult" age group and 70 instances of the "Senior" age group. The minimal misclassifications, with only 1 instance of "Senior" incorrectly predicted as "Adult" and 11 instances of "Adult" incorrectly predicted as "Senior," underscore the robustness of the SVM model. This high accuracy indicates the effectiveness of the SVM algorithm in discerning patterns within the features to accurately classify individuals into their respective age groups.
