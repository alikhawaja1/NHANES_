# LDA
In this exploration, Linear Discriminant Analysis (LDA) is employed to discern discriminative patterns among variables and optimize the classification of individuals based on their health attributes. LDA delves into the multivariate relationships within our dataset, aiming to maximize the separation between distinct age groups (seniors and adults) by leveraging variables such as blood glucose, insulin levels, cholesterol, and BMI.
```r
library(MASS)
```

```r
# Set seed for reproducibility
set.seed(123)

# Generate random indices for train and test split
indices <- sample(1:nrow(nhanes_data), 0.8 * nrow(nhanes_data))

# Create training and testing sets
train_data <- nhanes_data[indices, ]
test_data <- nhanes_data[-indices, ]

# Fit LDA model
lda_model <- lda(age_group ~ ., data = train_data)

# Make predictions on the test set
lda_predictions <- predict(lda_model, newdata = test_data)

# Create confusion matrix manually
conf_matrix_lda <- table(lda_predictions$class, test_data$age_group)
print(conf_matrix_lda)

# Calculate accuracy
accuracy_lda <- sum(diag(conf_matrix_lda)) / sum(conf_matrix_lda)
cat("Accuracy:", accuracy_lda, "\n")
```

The confusion matrix reveals the classification performance of a model designed to distinguish between adults and seniors based on the provided dataset. The matrix indicates that among 363 instances where the true age group was 'Adult,' the model correctly predicted 363 cases (True Positives). Additionally, for the 'Senior' age group, the model correctly classified 74 cases out of 86 (True Positives). However, there were 7 misclassifications of 'Adult' instances as 'Senior' (False Positives) and 12 misclassifications of 'Senior' instances as 'Adult' (False Negatives). The overall accuracy of the classification model is calculated as 95.83%, reflecting the proportion of correct predictions out of the total instances. This high accuracy suggests that the model effectively discriminates between adults and seniors based on the selected variables, demonstrating promising performance for age group classification. However, further evaluation, including consideration of precision, recall, and F1-score, would provide a more comprehensive understanding of the model's effectiveness.

# Machine Learning
In leveraging machine learning for this dataset, we harness algorithms to discern intricate patterns among health indicators. This analytical approach holds the promise of uncovering predictive relationships, facilitating age-group classification, and providing invaluable insights into the nuanced dynamics of health-related variables for targeted interventions and informed decision-making.

```r
# Load necessary libraries
library(e1071)
library(pROC)

nhanes_data$age_group <- as.factor(nhanes_data$age_group)

nhanes_data_selected <- nhanes_data

# Split the data into training and testing sets
set.seed(123)
indices <- sample(1:nrow(nhanes_data_selected), 0.8 * nrow(nhanes_data_selected))
train_data <- nhanes_data_selected[indices, ]
test_data <- nhanes_data_selected[-indices, ]

# Train SVM model
svm_model <- svm(age_group ~ ., data = train_data)

# Make predictions on the test set
svm_predictions <- predict(svm_model, newdata = test_data)

# Evaluate performance
conf_matrix_svm <- table(svm_predictions, test_data$age_group)
accuracy_svm <- sum(diag(conf_matrix_svm)) / sum(conf_matrix_svm)

# Print confusion matrix and accuracy
print(conf_matrix_svm)
cat("Accuracy:", accuracy_svm, "\n")

# ROC curve
svm_roc <- roc(test_data$age_group, as.numeric(svm_predictions))
plot(svm_roc, main = "ROC Curve for SVM", col = "blue", lwd = 2)

# K-fold Cross-Validation
set.seed(123)
folds <- cut(seq(1, nrow(nhanes_data_selected)), breaks = 5, labels = FALSE)
cv_svm <- lapply(1:5, function(i) {
  test_indices <- which(folds == i, arr.ind = TRUE)
  test_data <- nhanes_data_selected[test_indices, ]
  train_data <- nhanes_data_selected[-test_indices, ]
  
  svm_model <- svm(age_group ~ ., data = train_data)
  svm_predictions <- predict(svm_model, newdata = test_data)
  
  conf_matrix <- table(svm_predictions, test_data$age_group)
  accuracy <- sum(diag(conf_matrix)) / sum(conf_matrix)
  
  return(accuracy)
})

# Print cross-validated results
cat("Cross-validated Accuracy:", mean(unlist(cv_svm)), "\n")
```

The presented confusion matrix outlines the performance of a Support Vector Machine (SVM) model in classifying age groups (1: Adult, 2: Senior) based on the provided dataset. The matrix indicates that out of 388 instances where the true age group was 'Adult' (class 1), the SVM model correctly predicted 375 cases (True Positives), with 13 instances misclassified as 'Senior' (False Negatives). For 'Senior' instances (class 2), all 68 cases were correctly predicted (True Positives), with no misclassifications. The overall accuracy of the SVM model stands at 97.15%, denoting the proportion of correct predictions out of the total instances.

Additionally, the cross-validated accuracy is reported as 96.93%, affirming the robustness of the model across different subsets of the data. This high accuracy underscores the effectiveness of the SVM algorithm in distinguishing between adult and senior age groups based on the selected health variables. The model's minimal misclassifications suggest its reliability for age classification, offering valuable support for targeted health assessments and interventions in alignment with distinct age-related health dynamics.
