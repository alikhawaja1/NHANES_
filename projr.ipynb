```r
# Step 1: Install and load necessary libraries
install.packages("cellranger")
install.packages("pkgconfig")
install.packages("readxl")
library(readxl)

# Step 2: Load AHAH dataset into a data frame
# Assuming your dataset is in an Excel file named 'MANPO98.xlsx'
file_path <- '/users/banog/desktop/xx.xlsx'
ahah_df <- read_excel(file_path)

# Step 3: Exploratory Data Analysis (EDA)
# Display basic information about the dataset
str(ahah_df)

# Display summary statistics
summary(ahah_df)

for (column in c("ah3gp", "ah3dent", "ah3phar")) {
  hist(ahah_df[[column]], main = paste("Distribution of", column), col = "blue", xlab = column)
}

# Identify missing values
colSums(is.na(ahah_df))

# Subset LSOAs for England and Wales
england_wales_data <- ahah_df[grepl("^E", ahah_df$lsoa11), ]

# Subset LSOAs for Scotland
scotland_data <- ahah_df[grepl("^W", ahah_df$lsoa11), ]

# Example t-test for ah3gp variable
t_test_result <- t.test(england_wales_data$ah3gp, scotland_data$ah3gp)
print(t_test_result)
```

```r
# Visualization
# Create visual representations of key findings using base R plotting functions
# Example 1: Boxplot of ah3gp across regions
boxplot(ah3gp ~ substring(lsoa11, 1, 1), data = ahah_df, col = "lightblue", main = "Boxplot of ah3gp across Regions")

# Example 2: Histogram of ah3hosp
hist(ahah_df$ah3hosp, col = "salmon", main = "Histogram of ah3hosp", xlab = "ah3hosp")

# Example 3: Scatter plot between ah3ffood and ah3leis
plot(ahah_df$ah3ffood, ahah_df$ah3leis, col = "darkgreen", main = "Scatter Plot between ah3ffood and ah3leis", xlab = "ah3ffood", ylab = "ah3leis")

# Example 4: Bar plot of average ah3pubs by region
avg_pubs_by_region <- tapply(ahah_df$ah3pubs, substring(ahah_df$lsoa11, 1, 1), mean)
barplot(avg_pubs_by_region, col = "skyblue", main = "Average ah3pubs by Region", xlab = "Region", ylab = "Average ah3pubs")

# Example 5: Pie chart of ah3gamb percentages
pie(ahah_df$ah3gamb_pct, labels = ahah_df$lsoa11, main = "Pie Chart of ah3gamb Percentages")

# Example 6: Correlation heatmap for selected variables
cor_variables <- ahah_df[, c("ah3gp", "ah3dent", "ah3phar", "ah3hosp")]
cor_matrix <- cor(cor_variables)
heatmap(cor_matrix, col = colorRampPalette(c("white", "blue"))(20), main = "Correlation Heatmap")
# Example 7:
boxplot(ah3dent ~ substring(lsoa11, 1, 1), data = ahah_df, col = "lightblue", main = "Boxplot of ah3dent across Regions")

# Hypothesis Testing
# Formulate hypotheses and conduct hypothesis testing
# For example, testing if there is a significant difference in ah3phar between regions
t_test_result <- t.test(england_wales_data$ah3phar, scotland_data$ah3phar)
print(t_test_result)
# Example 2: T-test for ah3gp between England_Wales and Scotland
t_test_ah3gp <- t.test(england_wales_data$ah3gp, scotland_data$ah3gp)
print(t_test_ah3gp)

# Example 3: Wilcoxon test for ah3hosp between England_Wales and Scotland
wilcox_test_ah3hosp <- wilcox.test(england_wales_data$ah3hosp, scotland_data$ah3hosp)
print(wilcox_test_ah3hosp)

# Example 4: ANOVA for ah3leis among different regions
anova_ah3leis <- aov(ah3leis ~ substring(lsoa11, 1, 1), data = ahah_df)
summary(anova_ah3leis)

# Example 5: Chi-squared test for ah3gamb_pct and region
chi_squared_ah3gamb <- chisq.test(ahah_df$ah3gamb_pct, ahah_df$lsoa11)
print(chi_squared_ah3gamb)

# Example 6: Paired t-test for ah3ffood and ah3pubs
paired_t_test <- t.test(ahah_df$ah3ffood, ahah_df$ah3pubs, paired = TRUE)
print(paired_t_test)

# Example 7: Kruskal-Wallis test for ah3pm10 among different regions
kruskal_test_ah3pm10 <- kruskal.test(ah3pm10 ~ substring(lsoa11, 1, 1), data = ahah_df)
print(kruskal_test_ah3pm10)

```

1. **Welch Two Sample t-test (ah3phar):**
   The Welch Two Sample t-test was conducted to compare the means of ah3phar between Lower Level Super Output Areas (LSOAs) for England and Wales and Data Zones (DZ) for Scotland. The test yielded a t-value of -7.748 with 2041 degrees of freedom and a very low p-value (1.46e-14). This result indicates a statistically significant difference in the means of ah3phar between the two regions. The 95% confidence interval for the difference in means (-1.4875158 to -0.8865967) further supports this conclusion, as it does not include zero. The mean ah3phar in England/Wales (3.952814) is significantly lower than in Scotland (5.139871).

2. **Welch Two Sample t-test (ah3gp):**
   Similar to the ah3phar test, the Welch Two Sample t-test for ah3gp also revealed a significant difference in means between England/Wales and Scotland. The t-value was -10.299 with 2022.9 degrees of freedom and an extremely low p-value (< 2.2e-16). The 95% confidence interval for the difference in means (-1.962203 to -1.334460) supports the conclusion of a substantial difference. The mean ah3gp in England/Wales (4.657605) is significantly lower than in Scotland (6.305936).

3. **Wilcoxon Rank Sum Test (ah3hosp):**
   The Wilcoxon rank sum test was employed to assess the location shift in ah3hosp between England/Wales and Scotland. The test resulted in a W statistic of 9400386 and a p-value smaller than 2.2e-16, indicating a highly significant difference in location. This suggests that the distribution of ah3hosp values differs significantly between the two regions.

4. **Chi-squared Test (ah3gamb_pct):**
   The Pearson's Chi-squared test was used to examine the association between ah3gamb_pct and regions. The test returned an X-squared value of 4172900, with a large number of degrees of freedom (4172800) and a p-value of 0.4861. The p-value is greater than 0.05, suggesting that there is no significant association between ah3gamb_pct and regions.

5. **Paired t-test (ah3ffood vs. ah3pubs):**
   A paired t-test was conducted to compare the means of ah3ffood and ah3pubs within the same regions. The t-value was 30.944 with 41728 degrees of freedom and a p-value smaller than 2.2e-16. This indicates a highly significant difference in means, suggesting that the mean ah3ffood is significantly different from the mean ah3pubs within the same regions.

6. **Kruskal-Wallis Test (ah3pm10):**
   The Kruskal-Wallis test, a non-parametric test, was performed to compare the medians of ah3pm10 among different regions. The test resulted in a Kruskal-Wallis chi-squared value of 17798, with 2 degrees of freedom and a p-value smaller than 2.2e-16. This strongly indicates a significant difference in medians among the regions for ah3pm10.
